```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# STEP 1: Generate synthetic 2D data
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Plot original data (just for reference)
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Step 1: Unlabeled data (to be clustered)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
# STEP 2: Apply KMeans with K=3
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Get predicted labels and centroids
y_kmeans = kmeans.predict(X)
centroids = kmeans.cluster_centers_
```

```python
# STEP 3: Plot clustered data with centroids
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', edgecolor='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title("Step 3: K-Means Clustered Data with Centroids")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# STEP 4: Inertia = total within-cluster sum of square distances
print("Inertia (sum of distances to centroids):", kmeans.inertia_)
```
